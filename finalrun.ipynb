```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.patches as patches
import pydicom
import numpy as np
import cv2

def visualize_volume_results(result, n=3):
    targets = [r for r in result.top_suspicious if r.has_gradcam][:n]
    if not targets:
        print("No Grad-CAM overlays available.")
        return

    fig, axes = plt.subplots(len(targets), 3, figsize=(15, 5 * len(targets)))
    if len(targets) == 1:
        axes = [axes]

    for row, rec in enumerate(targets):

        # ── Load full DICOM slice ──────────────────────────────────
        ds = pydicom.dcmread(rec.file_path)
        hu = ds.pixel_array.astype(np.float32)
        slope = float(getattr(ds, "RescaleSlope", 1))
        intercept = float(getattr(ds, "RescaleIntercept", 0))
        hu = hu * slope + intercept

        # Lung window for display
        hu_display = np.clip(hu, -1350, 150)
        hu_display = ((hu_display + 1350) / 1500 * 255).astype(np.uint8)

        # ── Pull bbox, patch, Grad-CAM from pipeline result ───────
        pr = rec.pipeline_result
        bbox = pr.candidate_patches[0].bbox       # (x, y, w, h) in full-slice coords
        patch = pr.candidate_patches[0].patch     # raw cropped region
        overlay = pr.gradcam_overlays[0]          # Grad-CAM on 224×224 crop
        heatmap = pr.gradcam_heatmaps[0]

        prob = rec.primary_prediction.probability
        pred = rec.primary_prediction.prediction
        x, y, w, h = bbox

        # ── Col 0: Full slice + bounding box ──────────────────────
        axes[row][0].imshow(hu_display, cmap="gray")
        rect = patches.Rectangle(
            (x, y), w, h,
            linewidth=2, edgecolor="red", facecolor="none"
        )
        axes[row][0].add_patch(rect)
        axes[row][0].set_title(
            f"Full CT Slice — Slice {rec.slice_index:03d} | z={rec.z_position:+.1f}mm\n"
            f"Red box = ROI sent to model", fontsize=9
        )
        axes[row][0].axis("off")

        # ── Col 1: Cropped ROI patch (what the model saw) ─────────
        patch_display = np.clip(patch, -1350, 150)
        patch_display = ((patch_display + 1350) / 1500 * 255).astype(np.uint8)
        axes[row][1].imshow(patch_display, cmap="gray")
        axes[row][1].set_title(
            f"Cropped ROI ({w}×{h}px)\n"
            f"{pred} | prob={prob:.4f}", fontsize=9
        )
        axes[row][1].axis("off")

        # ── Col 2: Grad-CAM overlay on the cropped patch ──────────
        axes[row][2].imshow(overlay)
        # Draw the activation bbox on top
        heatmap_resized = cv2.resize(heatmap, (224, 224))
        heatmap_u8 = (heatmap_resized * 255).astype(np.uint8)
        _, binary = cv2.threshold(heatmap_u8, 100, 255, cv2.THRESH_BINARY)
        contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if contours:
            cx, cy, cw, ch = cv2.boundingRect(max(contours, key=cv2.contourArea))
            act_rect = patches.Rectangle(
                (cx, cy), cw, ch,
                linewidth=2, edgecolor="yellow", facecolor="none",
                label="Peak activation"
            )
            axes[row][2].add_patch(act_rect)
        axes[row][2].set_title(
            f"Grad-CAM on ROI\nYellow = peak activation region", fontsize=9
        )
        axes[row][2].axis("off")

    plt.suptitle(
        f"Patient: {result.patient_id} | Model: {result.model_name}\n"
        f"Volume: {result.volume_prediction} | prob={result.volume_probability:.4f} | {result.volume_confidence} confidence",
        fontsize=12, fontweight="bold", y=1.01,
    )
    plt.tight_layout()
    plt.savefig("volume_inference_visualization.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("Saved → volume_inference_visualization.png")

visualize_volume_results(result, n=3)
```
